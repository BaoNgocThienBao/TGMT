```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

#Img Path
_Path3 = "C:/Users/ADMIN/tgmt/image/hehe1.jpg"
#Open Img
image = Image.open(_Path3)
#Gray_Img
RIM =image.convert('L')
#Array Img
R_RIM=np.array(RIM)

#Binary image 
bR_RIM = 1*(R_RIM<200)

# Thực hiện opening để loại bỏ nhiễu và nối các đối tượng
opened_image = ndimage.binary_opening(bR_RIM, structure=np.ones((5,5)))

#Labeling
labeled_image, num_features = ndimage.label(bR_RIM)

# Tạo lưới subplot với 4 dòng và 1 cột
fig, axs = plt.subplots(4, 1, figsize=(30, 30))

plt.subplot(4,1,1)
plt.imshow(RIM, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(4,1,2)
plt.imshow(bR_RIM, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("Image after opening")

plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")

```

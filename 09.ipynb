```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

#Path
Dir_path='C:/Users/ADMIN/tgmt/image/zhezhi.jpg'
Dir_path2='C:/Users/ADMIN/tgmt/image/hehe1.jpg'

#Read
Read_img=Image.open(Dir_path)
# plt.imshow(Read_img)

#Convert -> gray
R_img_gray = Read_img.convert('L')
plt.imshow(R_img_gray,cmap='gray')

#Numberlize img
Img_array = np.array(R_img_gray)
print(Img_array)

#Create a new Fig
plt.figure()
plt.hist(Img_array.flatten(),bins=128)
plt.title('Histogram')
plt.ylabel('Quantity')
plt.xlabel('Value of pixel')
plt.show()

#Highline Linear
plt.figure()
plt.gray()
plt.contour(R_img_gray,origin='image')
plt.show()

#Redo
R_im=Image.open(Dir_path2).convert('L')
plt.imshow(R_im)
plt.figure()
plt.gray()
plt.contour(R_im,origin="image")
plt.axis('equal')
plt.show()
```

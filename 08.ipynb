```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud

#Merge line chart
x=[1,3,4,5,9]
y=[10,16,22,34,51]
x1=[2,5,7,8,13]
y1=[5,12,34,60,70]
plt.plot(x,y,'bo-',label="Dataset 1")
plt.plot(x1,y1,'go-',label="Dataset 2")
plt.legend()
plt.xlabel('X axis') 
plt.ylabel('Y axis')
plt.show()
```

```python
#Bar chart
Lib=['LOL','Val','CS2','Dota 2','?']
values =[25,17,13,14,30]
plt.bar(Lib,values,color='b',alpha=0.5)
plt.xlabel('Games')
plt.ylabel('Players')
plt.show()
```

```python
#Circle Chart
Actors=['LOL','Val','CS2','Dota 2','?']
pers =[25,17,13,30,15]
diff=['red','blue','yellow','green','purple']
plt.pie(pers, labels=Actors, colors = diff, autopct='%1.1f%%')
plt.title('Circle Chart')
plt.show()
```

```python
#Dot chart
x=[1,3,4,5,9]
y=[10,16,22,34,51]
plt.scatter(x,y,color='r',label='Dataset')
plt.title('Dot chart')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#Box chart
dataset=[25,17,13,14,30,22,31,27]
plt.boxplot(dataset)
plt.title('Box chart')
plt.ylabel('Y')
plt.show()
```

```python
#Expand chart
dataset=[25,17,13,14,30,22,31,27]
plt.violinplot(dataset)
plt.title('Violin Chart')
plt.ylabel('Y')
plt.show()
```

```python
#Word Count
#!pip install wordcloud (If you haven't install this Lib yet)

text="Today is the day that i have been blessed by a beautiful angel i've just met at my school"
wordcloud = WordCloud(width=800 , height=400).generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.title('Word Count')
plt.show()
```

```python
#Radar Chart
Lib1=['LOL','Val','CS2','Dota 2','?']
values=[1,3,4,5,9]
N = len(Lib1)
values += values[:1]
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]
plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Radar chart')
plt.show()
```

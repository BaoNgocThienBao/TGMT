```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

#Img Path
_Path1 = 'C:/Users/ADMIN/tgmt/image/zhezhi.jpg'
#Open Img
image = Image.open(_Path1)
#Gray_Img
RIM =image.convert('L')
#Array Img
R_RIM=np.array(RIM)
#Blur Img
BR_RIM_1 = ndimage.gaussian_filter(R_RIM,sigma=1,mode="constant")
BR_RIM_2 = ndimage.gaussian_filter(R_RIM,sigma=8,mode="constant")
BR_RIM_3 = ndimage.gaussian_filter(R_RIM,sigma=15,mode="constant")

    #Display Img
fig, axs = plt.subplots(2,2, figsize=(10, 8))

#Origin gray image
plt.subplot(2,2,1)
plt.imshow(RIM,cmap="gray")
plt.title("Origin")

#Blur 1
plt.subplot(2,2,2)
plt.imshow(BR_RIM_1,cmap="gray")
plt.title("Blur, Sigma 1")

#Blur 2
plt.subplot(2,2,3)
plt.imshow(BR_RIM_2,cmap="gray")
plt.title("Blur, Sigma 8")

#Blur 3
plt.subplot(2,2,4)
plt.imshow(BR_RIM_3,cmap="gray")
plt.title("Blur, Sigma 15")
```

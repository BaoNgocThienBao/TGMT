```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from def_for_13_14 import histogram_equalization

_Path = 'C:/Users/ADMIN/tgmt/image/zhezhi.jpg'
Rim = Image.open(_Path)
Rim_GRAY = Rim.convert('L')
Rim_AR_GRAY= np.array(Rim_GRAY)

# Use Balancing Histogram Gray
equalized_image = histogram_equalization(Rim)

# Create a 2x2 web to display image and histogram
plt.figure(figsize=(15, 10))

# Show origin image
plt.subplot(2, 2, 1)
plt.imshow(Rim_GRAY, cmap='gray')
plt.title('Hình ảnh gốc')

# Draw histogram of origin image
plt.subplot(2, 2, 2)
plt.hist(Rim_AR_GRAY.flatten(), bins=128)
plt.title('Biểu đồ Histogram - Hình ảnh gốc')

# Display image being equalized
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Hình ảnh đã cân bằng')

# Histogram of equalized image
plt.subplot(2, 2, 4)
plt.hist(np.array(equalized_image).flatten(), bins=128)
plt.title('Biểu đồ Histogram - Hình ảnh đã cân bằng')

plt.show()
```

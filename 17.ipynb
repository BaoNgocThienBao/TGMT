```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

#Img Path
_Path1 = 'C:/Users/ADMIN/tgmt/image/zhezhi.jpg'
#Open Img
image = Image.open(_Path1)
#Gray_Img
RIM =image.convert('L')
#Array Img
R_RIM=np.array(RIM)

# Calculate gradient base on x axis 
gradient_x0 = ndimage.sobel(R_RIM, axis=0, mode='constant')
gradient_x1 = ndimage.sobel(R_RIM, axis=1, mode='constant')

# Tính gradient theo hướng x (đạo hàm riêng theo x) bằng mặt nạ Prewitt x
gradient_x = ndimage.convolve(R_RIM, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tính gradient theo hướng y (đạo hàm riêng theo y) bằng mặt nạ Prewitt y
gradient_y = ndimage.convolve(R_RIM, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tính biên độ gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Display img
fig, axs = plt.subplots(2,2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(R_RIM, cmap='gray')
plt.title('Origin gray image')

plt.subplot(2,2,2)
plt.imshow(gradient_x0, cmap='gray')
plt.title('Ảnh xám x=0')

plt.subplot(2,2,3)
plt.imshow(gradient_x1, cmap='gray')
plt.title('Ảnh xám x=1')

plt.subplot(2,2,4)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Ảnh gradient magnitude')
plt.show()
```
